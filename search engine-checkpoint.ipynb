```python
from gensim.models import Word2Vec
from langchain_community.vectorstores.chroma import Chroma

# Load the Word2Vec model
word2vec_model = Word2Vec.load("skipgram_model1.bin")

# Define a function to generate document embeddings
# Define a function to generate document embeddings
def generate_document_embeddings(documents, word2vec_model):
    document_embeddings = []
    for document in documents:
        # Extract text content from the Document object
        text_content = document.page_content
        
        # Tokenize the document into words
        words = text_content.split()  # You might need a more sophisticated tokenizer here
        
        # Generate embeddings for each word in the document
        word_embeddings = []
        for word in words:
            if word in word2vec_model.wv:
                word_embeddings.append(word2vec_model.wv[word])
        
        # Average the word embeddings to get a single document embedding
        if word_embeddings:
            document_embedding = sum(word_embeddings) / len(word_embeddings)
            document_embeddings.append(document_embedding)
        else:
            # If no word embeddings were found for the document, append None
            document_embeddings.append(None)
    
    return document_embeddings


# Assuming 'documents' is a list containing your text data
# Generate document embeddings
# Create a new instance of Chroma
vectordb = Chroma(persist_directory=persist_directory)

# Add the document embeddings to Chroma
for i, embedding in enumerate(document_embeddings):
    if embedding is not None:
        # Add the embedding to Chroma with an identifier (e.g., document ID)
        vectordb.add_vector(str(i), embedding)

# Optionally, you can persist the Chroma instance to disk
vectordb.persist()
```

```python
def retrieve_similar_documents(query_embedding, vectordb, top_k=5):
    """
    Retrieve similar documents to the given query embedding from the Chroma vector database.
    
    Args:
        query_embedding (numpy.ndarray): The embedding of the query document.
        vectordb (Chroma): The Chroma vector database.
        top_k (int): Number of similar documents to retrieve.

    Returns:
        List of tuples (document_id, similarity_score) for the top_k similar documents.
    """
    # Retrieve similar embeddings from the Chroma database
    similar_embeddings = vectordb.retrieve_similar(query_embedding, k=top_k)
    
    # Convert similar embeddings to document IDs and similarity scores
    similar_documents = [(doc_id, similarity) for doc_id, similarity, _ in similar_embeddings]
    
    return similar_documents
```

```python
# Assuming you have a query_embedding and a Chroma instance named vectordb
similar_documents = retrieve_similar_documents(query_embedding, vectordb, top_k=5)

# Print the top similar documents
for doc_id, similarity_score in similar_documents:
    print(f"Document ID: {doc_id}, Similarity Score: {similarity_score}")
```
